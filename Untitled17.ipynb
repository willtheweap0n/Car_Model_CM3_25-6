```python
# STEP 1: Get high-resolution road route from OSRM
import requests
import pandas as pd
import time

# Start and end points: [lon, lat]
start = [-3.1883, 55.9533]  # Edinburgh
end = [-0.1276, 51.5074]    # London

# Request route from OSRM
osrm_url = f"https://router.project-osrm.org/route/v1/driving/{start[0]},{start[1]};{end[0]},{end[1]}?overview=full&geometries=geojson"
r = requests.get(osrm_url)

print("OSRM Status:", r.status_code)
osrm_data = r.json()

# Extract coordinates
coords = osrm_data["routes"][0]["geometry"]["coordinates"]

# Convert to DataFrame
df = pd.DataFrame(coords, columns=["lon", "lat"])
print(f"Received {len(df)} route points from OSRM.")
df.head()

# STEP 2: Add elevation using ORS Elevation API

api_key = "YOUR_ORS_KEY_HERE"  # your existing key is fine

def get_elevations(coords, api_key):
    """Query ORS elevation API in batches of 100 points."""
    all_elevs = []
    for i in range(0, len(coords), 100):
        batch = coords[i:i+100]
        locations = [{"latitude": lat, "longitude": lon} for lon, lat in batch]
        body = {"format_in": "point", "format_out": "point", "points": locations}
        headers = {"Authorization": api_key, "Content-Type": "application/json"}
        url = "https://api.openrouteservice.org/elevation/point"
        res = requests.post(url, headers=headers, json=body)
        if res.status_code == 200:
            data = res.json()
            for p in data["geometry"]["coordinates"]:
                all_elevs.append(p[2])
        else:
            print("Error:", res.status_code, res.text)
            break
        time.sleep(0.2)
    return all_elevs

coords_list = df[["lon", "lat"]].values.tolist()
elevations = get_elevations(coords_list, api_key)

if len(elevations) == len(df):
    df["elev_m"] = elevations
else:
    print("Warning: some elevation points missing.")
    df["elev_m"] = elevations + [None]*(len(df)-len(elevations))

# Save final dataset
df.to_csv("osrm_ors_route_highres.csv", index=False)
print(f"\n✅ Saved {len(df)} route points with elevation to 'osrm_ors_route_highres.csv'.")
print(df.head())
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree

# Load both datasets
df_dense = pd.read_csv("osrm_ors_route_highres.csv")   # many points, no elevation
df_sparse = pd.read_csv("ors_route_fastest.csv")        # fewer points, with elevation

# Keep only points that have elevation
df_sparse = df_sparse.dropna(subset=["elev_m"])

# Build a KD-tree for fast nearest-neighbor search
tree = cKDTree(df_sparse[["lon", "lat"]].values)

# For each dense point, find the nearest sparse (elevation) point
distances, indices = tree.query(df_dense[["lon", "lat"]].values, k=1)

# Assign interpolated elevation
df_dense["elev_m"] = df_sparse.iloc[indices]["elev_m"].values

# Optional: smooth elevations to remove step changes (simple moving average)
df_dense["elev_m"] = df_dense["elev_m"].rolling(window=5, center=True, min_periods=1).mean()

# Save the combined high-resolution dataset
df_dense.to_csv("final_route_highres_elevation.csv", index=False)

print(f"✅ Combined dataset saved with {len(df_dense)} points.")
print(df_dense.head())

```

```python
import numpy as np
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Load your route data (has lon, lat)
df = pd.read_csv("final_route_highres_elevation.csv")

# Convert to radians
R = 6371000  # Earth radius (m)
lat = np.radians(df["lat"].to_numpy())
lon = np.radians(df["lon"].to_numpy())

# Compute distances between successive points
dlat = lat[1:] - lat[:-1]
dlon = lon[1:] - lon[:-1]
a = np.sin(dlat/2)**2 + np.cos(lat[:-1]) * np.cos(lat[1:]) * np.sin(dlon/2)**2
c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
distances = R * c  # metres

# Add to DataFrame
df["segment_length_m"] = np.concatenate([[0], distances])
df["cum_dist_m"] = np.cumsum(df["segment_length_m"])

print(df.head())
print(f"Average point spacing: {np.mean(distances):.1f} m")
```

```python
import osmnx as ox
import pandas as pd
import numpy as np
from shapely.geometry import Point

# Load your route file (must exist in your working directory)
df = pd.read_csv("final_route_highres_elevation.csv")

# Define a small bounding box around your route
buffer_deg = 0.05
north = df["lat"].max() + buffer_deg
south = df["lat"].min() - buffer_deg
east  = df["lon"].max() + buffer_deg
west  = df["lon"].min() - buffer_deg

print("📡 Downloading road network (this may take a few minutes)...")

# ✅ Correct syntax for OSMnx 2.0.6 (keyword-only)
G = ox.graph_from_bbox(north=north, south=south, east=east, west=west, network_type='drive')

# Add estimated speeds (from OSM data or typical highway types)
G = ox.add_edge_speeds(G)

# Add estimated travel times based on distance and speed
G = ox.add_edge_travel_times(G)

# Match each route point to its nearest OSM node
nodes = ox.distance.nearest_nodes(G, X=df["lon"].values, Y=df["lat"].values)

# Get edge speed (kph) for each segment between points
edge_speeds = []
for i in range(len(nodes) - 1):
    try:
        edges = G.get_edge_data(nodes[i], nodes[i+1])
        if edges:
            data = min(edges.values(), key=lambda x: x.get("travel_time", np.inf))
            edge_speeds.append(data.get("speed_kph", np.nan))
        else:
            edge_speeds.append(np.nan)
    except Exception:
        edge_speeds.append(np.nan)

# Pad the last point so arrays align
edge_speeds.append(edge_speeds[-1])

# Add to DataFrame
df["speed_limit_kph"] = edge_speeds

# Save the result
df.to_csv("route_with_speed_limits.csv", index=False)
print(f"✅ Saved {len(df)} points with speed limits to 'route_with_speed_limits.csv'")
print(df.head())

```

```python
import osmnx as ox
!pip install osmnx geopandas shapely

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Load your route file (the one with lon, lat, elev_m)
df = pd.read_csv("final_route_highres_elevation.csv")

# Convert to GeoDataFrame for spatial operations
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs="EPSG:4326")

# Define bounding box + small buffer
buffer_deg = 0.05
north, south = df["lat"].max() + buffer_deg, df["lat"].min() - buffer_deg
east, west   = df["lon"].max() + buffer_deg, df["lon"].min() - buffer_deg

print("📡 Downloading road data from OpenStreetMap...")
# Get all road geometries that have a 'maxspeed' tag
roads = ox.geometries_from_bbox(north, south, east, west, tags={"highway": True, "maxspeed": True})

# Keep only useful columns
roads = roads[["highway", "maxspeed", "geometry"]].dropna(subset=["maxspeed"])
print(f"✅ Downloaded {len(roads)} road segments with speed limits.")

# Ensure both are in same coordinate reference system
roads = roads.to_crs(epsg=4326)

# Find nearest road for each route point
print("🔍 Matching route points to nearest roads...")
joined = gpd.sjoin_nearest(gdf, roads, how="left", distance_col="dist_to_road")

# Clean up and save results
joined = joined.rename(columns={"maxspeed": "speed_limit"})
joined = joined.drop(columns=["index_right"])
joined.to_csv("route_with_osm_speed_limits.csv", index=False)

print(f"✅ Saved {len(joined)} points with OSM speed limits to 'route_with_osm_speed_limits.csv'")
print(joined[["lon", "lat", "elev_m", "speed_limit", "dist_to_road"]].head())

```
